# Parsing CSV from yield curve html

```python
pattern = r"(?:<td.*?>(.*?)</td>)|(N/A)"
import re
```

## data shape

```python
data = {'label':['Date','1 mo','2 mo','3 mo','6 mo','1 yr','2 yr','3 yr','5 yr','7 yr','10 yr','20 yr','30 yr'],'rows':[]}
```

```python
l = []
```

```python
with open('yield_curve_table.html') as html:
    i = 0
    temp = []
    for line in html:
        for pair in re.findall(pattern,line):
            for item in pair:
                if item != '':
                    temp.append(item)
                    i+=1
                    if i==13:
                        i=0
                        l.append(temp)
                        temp = []
```

```python
data['rows'] = l
```

```python
data
```

## writing it down

```python
with open('yield_curve_daily_data.csv' , 'w') as file:
    file.write(','.join(data['label'])+'\n')
    for row in data['rows']:
        file.write(','.join(row)+'\n')
```
