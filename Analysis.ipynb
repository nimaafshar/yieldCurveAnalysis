```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('yield_curve_daily_data.csv')
```

```python
data.tail(20)
```

```python
data['month'] = data['Date'].apply(lambda x:x.split('/')[0])
data['day'] = data['Date'].apply(lambda x:x.split('/')[1])
data['year'] = data['Date'].apply(lambda x:x.split('/')[2])
```

```python
grouped_by_month = data.groupby(['year','month'])
```

```python
grouped_by_year = data.groupby('year')
```

```python
#fixing year
data['year'] = data.year.apply(lambda year:'19'+year if int(year)> 89 else '20'+year)
```

```python
mean_by_month = grouped_by_month.mean()
```

```python
mean_by_year = grouped_by_year.mean()
```

```python
def plot_year(year):
    plt.plot(keys,means_by_year.iloc[int(year)-1990].values[5:10],'b',linewidth=5.0)
    plt.title(year)
    plt.show()
```

```python
#finding min & max
mins = []
maxs = []
for key in data.keys()[6:11]:
    mins.append(data[key].min())
    maxs.append(data[key].max())
max_ = max(maxs)
min_ = min(mins)
```

```python
fig ,ax = plt.subplots(5,6)
for i in range(5):
    for j in range(6):
        index = 6*i+j
        color = 'b' if means_by_year.iloc[index].values[9] - means_by_year.iloc[index].values[5] >0 else 'r'
        ax[i,j].plot(keys,means_by_year.iloc[index].values[5:10],color,linewidth=1.0)
        ax[i,j].set_title(1990+index)
        ax[i,j].set_ylim([min_, max_])
        
fig.set_size_inches(18.5, 10.5, forward=True)
fig.tight_layout()
plt.show()
```

```python
month_names = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']
```

```python
def draw_for_year(year,save_to=None,show=True):
    fig ,ax = plt.subplots(4,3)
    for i in range(4):
        for j in range(3):
            month_number = 3*i+j 
            record = mean_by_month[(mean_by_month.year==year) & (mean_by_month.month == month_number+1)]
            if len(record) > 0:
                values = record.values[0][7:12]
                color = 'b' if values[-1] - values[0] >0 else 'r'
                ax[i,j].plot(keys,values,color,linewidth=1.0)
                ax[i,j].set_title(month_names[month_number])
                ax[i,j].set_ylim([min_, max_])
            else:
                ax[i,j].axis('off')
    fig.set_size_inches(8, 6, forward=True)
    fig.tight_layout()
    fig.suptitle(year,fontsize=15)
    plt.subplots_adjust(top=0.9)
    if save_to!=None : plt.savefig(save_to)
    if show: plt.show()
```

```python
mean_by_month['year'] = pd.to_numeric(mean_by_month['year'])
mean_by_month['month'] = pd.to_numeric(mean_by_month['month'])
```

```python
diff_data = data[['10 yr','2 yr','Date']]
```

```python
diff_data['diff'] = diff_data['10 yr'] - diff_data['2 yr']
```

```python
def datestring_to_float(date):
    month , day , year = map(int ,date.split('/'))
    year = year+1900 if year > 50 else year+2000
    return year + (month-1)*(1/12) + (day-1)*(1/(12*30))
```

```python
plt.plot(diff_data['Date'].apply(datestring_to_float),diff_data['diff'])
plt.axhline(y=0.0, color='r', linestyle='-')
plt.show()
```
